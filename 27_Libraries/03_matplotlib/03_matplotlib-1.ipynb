# Matplotlib Basics

```python
%matplotlib inline
from matplotlib import pyplot as plt
```

```python
plt.plot([1,2,3],[2,5,7])   # creates a line graph through (1,2) (2,5) (3,7)
plt.title("Line Graph")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.tight_layout()      # improves padding
plt.grid(True)      # grid
plt.show()
```

```python
print(plt.style.available)
```

```python
from matplotlib import style
style.use('ggplot')     # new style
x = [1,5,7]
y = [23, 45, 89]
x1 = [3, 7, 3]
y1 = [5,8,9]
```

```python
plt.hist(x,label="First",linewidth=3,color="green")  # histogram plot
plt.legend()    # gives labels
plt.show()
```

```python
plt.scatter(x,y,label="Second",color="orange")  # scatter plot
plt.legend()    # gives labels
plt.show()
```

```python
plt.plot(x1,y1,label="Third",linewidth=1,marker="o",color="blue") # line plot
plt.legend()    # gives labels
plt.show()
```

```python
plt.plot(x,y1,label="Fourth",color="black",marker=".",linestyle="--") # -- plot
plt.legend()    # gives labels
plt.show()
```
