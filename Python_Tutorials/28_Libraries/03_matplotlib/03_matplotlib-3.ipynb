# Pie Chart

```python
from matplotlib import pyplot as plt
plt.style.use('fivethirtyeight')
```

```python
slices = [198, 154, 786, 1000, 486, 600,834]
labels = ["python","ruby","java","js","c","c++","PHP"]
explode = [0,0,0,0.1,0,0,0.3]
plt.pie(slices,labels=labels,explode=explode,shadow=True,startangle=90,autopct='%1.1f%%',radius=5,wedgeprops={'edgecolor':'black'})
plt.title('My Pie Chart')
```
